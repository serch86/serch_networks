```python
import networkx as nx
import numpy as np

import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
G.add_nodes_from(['a','b','c','i','j','k','d','e'])
```

```python
G.add_edges_from([('j','c'),('j','b'),('b','c'),('a','c'),('b','a'),
                  ('i','a'),('i','d'),('d','k'),('k','e'),('e','d')])
```

```python
nx.draw_networkx(G)
nx.betweenness_centrality(G)
plt.axis('off');
```

```python
nx.betweenness_centrality(G)
```

# Generemos redes llamativas
Comenzamos por las redes *Clasicas*

## Arbol balanceado

```python
H = nx.balanced_tree(3,2)
nx.draw_networkx(H)
plt.axis("off")
```

## Barbell Graph

```python
H = nx.barbell_graph(5,1)
nx.draw_networkx(H)
plt.axis("off")
```

## Comple Graph

```python
H = nx.complete_graph(8)
nx.draw_networkx(H)
plt.axis("off");
```

## Cycle Graph

```python
H = nx.cycle_graph(8)
nx.draw_networkx(H)
plt.axis("off");
```

## Dorogovtsev-Goltsev-Mendes graph

```python
H = nx.dorogovtsev_goltsev_mendes_graph(3)
nx.draw_networkx(H)
plt.axis("off");
```

## Grid Graph

```python
H = nx.grid_graph([5,5])
nx.draw_networkx(H)
plt.axis("off");
```

## Hypergrid Graph

```python
H = nx.hypercube_graph(4)
nx.draw_networkx(H,pos=nx.spring_layout(H))
plt.axis("off");
```

## Ladder Graph

```python
H = nx.ladder_graph(8)
nx.draw_networkx(H)
plt.axis("off");
```
